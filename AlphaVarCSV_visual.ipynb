```python
import numpy as np
import pandas as pd
import os
import glob
```

```python
path = 'Alpha_var_250img/CSV/'
# Get the files from the path provided in the OP
# files = glob.glob(os.path.join(path, "*.csv"))  # .rglob to get subdirectories
```

```python
real_loss = np.zeros((2000,1))
df = pd.read_csv(path+'Real_loss.csv')
df_numpy = df.to_numpy()
real_loss[0] = np.array(df.columns)
real_loss[1:] = df_numpy
```

```python
synth_loss = np.zeros((2000,1))
df = pd.read_csv(path+'Synth_loss.csv')
df_numpy = df.to_numpy()
synth_loss[0] = np.array(df.columns)
synth_loss[1:] = df_numpy
```

```python
plt.plot(real_loss)
plt.plot(synth_loss)
plt.title(' Real-blue vs Synth-Orange loss')
```
